```python
import pandas as pd

import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

**P1**

```python
dane = {
    "Rok": [2002,2011,2015,2022],
    "Warszawa": [1672000,1707000,1740000,1795000],
    "Łódź": [789318, 728892,None,670642]
}
```

**P2**

```python
df = pd.DataFrame(dane)
```

**P3**

```python
df.to_csv("populacja_miast.csv", index=False)
```

**P5**

```python
#PLT

plt.plot(df['Rok'], df['Warszawa'], label='Warszawa', color='blue', marker='o')
plt.plot(df['Rok'], df['Łódź'], label='Łódź', color='red', marker='x')
plt.title('Zmiany populacji Warszawy i Łodzi')
plt.xlabel('Rok')
plt.ylabel('Liczba mieszkańców')
plt.legend()
plt.show()
```

```python
#Pandas

df.set_index('Rok')[['Warszawa', 'Łódź']].plot(marker='o')
plt.title('Zmiany populacji Warszawy i Łodzi')
plt.xlabel('Rok')
plt.ylabel('Liczba mieszkańców')
plt.show()
```

```python
#seaborn

sns.lineplot(data=df, x='Rok', y='Warszawa', label='Warszawa', marker='o')
sns.lineplot(data=df, x='Rok', y='Łódź', label='Łódź', marker='x')
plt.title('Zmiany populacji Warszawy i Łodzi')
plt.xlabel('Rok')
plt.ylabel('Liczba mieszkańców')
plt.legend()
plt.show()
```

# **Zad. 2**

**P1**

```python
np.random.seed(42)
godziny_nauki = np.random.uniform(0, 10, 100)
szum = np.random.randint(-10, 10, 100)
ocena = 50 + godziny_nauki * 5 + szum
```

**P2**

```python
df_scat = pd.DataFrame({
    'godziny_nauki': godziny_nauki,
    'ocena': ocena
})
```

 **P3 & P5**

```python
plt.scatter(df_scat['godziny_nauki'], df_scat['ocena'], alpha=0.6)
plt.title('Zależność między godzinami nauki a oceną')
plt.xlabel('Godziny nauki')
plt.ylabel('Ocena')
plt.show()
```

**P4 & P5**

```python
sns.scatterplot(data=df_scat, x='godziny_nauki', y='ocena', alpha=0.6,color='purple')
sns.regplot(x='godziny_nauki', y='ocena', data=df_scat, scatter=False, color='red')
plt.title('Zależność między godzinami nauki a oceną')
plt.xlabel('Godziny nauki')
plt.ylabel('Ocena')
plt.show()
```

# Zad 3

**P1**

```python
wynagrodzenia = {
    'Miasto': ['Warszawa', 'Kraków', 'Wrocław', 'Poznań', 'Gdańsk', 'Łódź'],
    'Wynagrodzenie [PLN]': [8500, 7200, 7400, 7100, 6900, 6300]
}

df_wynagrodzenia = pd.DataFrame(wynagrodzenia)
df_wynagrodzenia
```

**P2, P4, P5**

```python
plt.barh(df_wynagrodzenia['Miasto'], df_wynagrodzenia['Wynagrodzenie [PLN]'], color=plt.cm.RdYlGn(df_wynagrodzenia['Wynagrodzenie [PLN]'] / df_wynagrodzenia['Wynagrodzenie [PLN]'].max()))
for i, v in enumerate(df_wynagrodzenia['Wynagrodzenie [PLN]']):
    plt.text(v + 100, i, str(v), va='center', color='black', fontsize=12)
plt.title('Przeciętne miesięczne wynagrodzenie w miastach Polski (2023)')
plt.xlabel('Wynagrodzenie [PLN]')
plt.show()
```

```python
sns.barplot(data=df_wynagrodzenia, y='Miasto', x='Wynagrodzenie [PLN]', palette='RdYlGn')
for index, value in enumerate(df_wynagrodzenia['Wynagrodzenie [PLN]']):
    plt.text(value + 100, index, str(value), va='center', color='black', fontsize=12)
plt.title('Przeciętne miesięczne wynagrodzenie w miastach Polski (2023)')
plt.xlabel('Wynagrodzenie [PLN]')
plt.show()
```
