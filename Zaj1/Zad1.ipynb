# **ZADANIE 3**

**Część 1**

```python
def calculate_mean(np_a):
  return __builtins__.sum(np_a)/len(np_a)

def calculate_median(np_a):
  sorted_numbers = sorted(np_a)
  n = len(sorted_numbers)
  if n % 2 == 1:
    return sorted_numbers[n // 2]
  else:
    mid1 = sorted_numbers[n // 2 - 1]
    mid2 = sorted_numbers[n // 2]
    return (mid1 + mid2) / 2
```

```python
np_a = [1,3,5,6,4,3,7,4,5,7]
print('Średnia =',calculate_mean(np_a))
print('Mediana =',calculate_median(np_a))
```

**Część 2**

```python
print("Średnia",np.mean(np_a))
print("Mediana",np.median(np_a))
```

```python
x = np.linspace(-2 * np.pi, 2 * np.pi, 500)
y1 = np.sin(x)
y2 = 2 * np.sin(x)

fig, (ax1, ax2) = plt.pyplot.subplots(1, 2, figsize=(10, 5))

ax1.plot(x, y1, color='blue')
ax1.set_title('y = sin(x)')
ax1.set_ylim(-2.1, 2.1)

ax2.plot(x, y2, color='red')
ax2.set_title('y = 2*sin(x)')
ax2.set_ylim(-2.1, 2.1)

plt.pyplot.show()
```

# **ZADANIE 5**

```python
def rozklad(x, mu, sigma):
    return (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-((x - mu)**2) / (2 * sigma**2))

x = np.linspace(-10, 10, 1000)

params = [
    (0, 1),
    (0, 2),
    (2, 1)
]

plt.pyplot.figure(figsize=(10, 6))

for mu, sigma in params:
    y = rozklad(x, mu, sigma)
    plt.pyplot.plot(x, y, label=f'μ = {mu}, σ = {sigma}')

plt.pyplot.title('Funkcja gęstości prawdopodobieństwa rozkładu normalnego')
plt.pyplot.xlabel('x')
plt.pyplot.ylabel('f(x)')
plt.pyplot.legend()
plt.pyplot.show()
```

# **ZADANIE 6**

```python
data1 = np.random.normal(0, 1, 1000)
data2 = np.random.normal(0, 2, 1000)
data3 = np.random.normal(2, 1, 1000)

plt.pyplot.figure(figsize=(10, 6))

plt.pyplot.hist(data1, bins=50, alpha=0.6, color='blue', label='μ=0, σ=1', density=True)
plt.pyplot.hist(data2, bins=50, alpha=0.6, color='green', label='μ=0, σ=2', density=True)
plt.pyplot.hist(data3, bins=50, alpha=0.6, color='red', label='μ=2, σ=1', density=True)

plt.pyplot.title('Histogramy rozkładów normalnych dla różnych μ i σ')
plt.pyplot.xlabel('Wartości')
plt.pyplot.ylabel('Gęstość prawdopodobieństwa')
plt.pyplot.legend()
plt.pyplot.show()
```

# **ZADANIE 7**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

params = [(0, 1), (0, 2), (2, 1)]

def rozklad(x, mu, sigma):
    return norm.pdf(x, mu, sigma)

n = 10000
x = np.linspace(-10, 10, 1000)

fig, axes = plt.subplots(3, 1, figsize=(3, 6))
colors_tab = ['blue', 'orange', 'green']

for i, (mu, sigma) in enumerate(params):
    ax = axes[i]
    data = np.random.normal(mu, sigma, n)

    ax.hist(data, bins=50, density=True, alpha=0.6, color=colors_tab[i], label='Dane losowe')
    y = rozklad(x, mu, sigma)
    ax.plot(x, y, linestyle='--', linewidth=1, color='red', label='Analityczna gęstość')


    ax.set_title(f'μ={mu}, σ={sigma}')
    ax.set_xlabel('x')
    ax.set_xlim(-9, 9)
    ax.set_ylabel('Gęstość  \n prawdopodobieństwa')
    ax.set_ylim(0, 0.4)

plt.tight_layout()
plt.show()
```
