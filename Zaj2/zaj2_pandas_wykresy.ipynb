**Zadanie 1**

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data = (
    {"rok": [2002, 2011, 2015, 2022],
     "Warszawa" : [1672000, 1707000, 1740000, 1795000],
     "Łódź" : [789318, 728892,"Brak danych" ,670642]
     }
)
mieszkancy=pd.DataFrame(data)
mieszkancy
```

```python
mieszkancy.to_csv('populacja_miast.csv', index=False)
```

**Zadanie 2**

```python
mieszkancy_df = pd.read_csv('populacja_miast.csv')
mieszkancy_df
```

```python

mieszkancy_df['Łódź'] = pd.to_numeric(mieszkancy_df['Łódź'], errors='coerce') # Convert 'Łódź' to numeric, coercing errors
mieszkancy_df.plot(x='rok', y=['Warszawa', 'Łódź'], kind='line', marker='o')
plt.title('Zmiana populacji Warszawy i Łodzi na przestrzeni lat')
plt.xlabel('Rok')
plt.ylabel('Populacja')
plt.grid(True)
plt.show()
```

```python
lata = mieszkancy_df[mieszkancy_df['rok'].isin([2002, 2011, 2022])]
lata.plot(x='rok', y=['Warszawa', 'Łódź'], kind='line', marker='o')
plt.title('Zmiana populacji Warszawy i Łodzi w latach 2002,2011,2022')
plt.xlabel('Rok')
plt.ylabel('Populacja')
plt.grid(True)
plt.show()
```
