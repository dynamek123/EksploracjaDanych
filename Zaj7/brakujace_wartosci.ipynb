```python
import numpy as np
import pandas as pd
import google.colab.files as files
uploaded = files.upload()
df = pd.read_csv('nieruchomosci.csv')
print(df.dtypes)

df.replace(['NA', 'N/A', 'brak','0','-1','-999', 'brak danych', '?', '-',-999,-1], np.nan, inplace=True)
#print(df.liczba_pokoi<1)
#print(df.cena_m2<1000)
#print(df.cena_total<1000)
#print(df.pietro<0)
#print(df.liczba_pieter<1)
#print(df.isnull())
#powyższe print sprawdzały mi czy są nieprawidłowe dane

df["stan"] = df.stan.fillna(df.stan.mode()[0])
df.loc[1, "stan"] = "Bardzo dobry"
#powyższe pole jest wpisane po chińsku jako Bardzo dobry wiec zmienilem je recznie
df["parking"] = df.parking.fillna(df.parking.mode()[0])
df["balkon"] = df.balkon.fillna(df.balkon.mode()[0])
#
#powyzsze sa jako object i są czysto tekstowe więc uzylem wypelnienia mode
#
df["pietro"] = df["pietro"].fillna(0)
df["rok_budowy"] = df.rok_budowy.fillna(df.rok_budowy.mode()[0])
df["powierzchnia"] = df.powierzchnia.fillna(df.powierzchnia.mode()[0])
#
#powyzsze 3 sa jako object a nie int wiec uzylem mode, natomiast dla pietra wypelnilem NaN jako 0 (pietro 0)
#
df["powierzchnia"] = pd.to_numeric(df["powierzchnia"], errors="coerce")
df["cena_m2"] = df.cena_m2.fillna(df.cena_m2.mean())
df["cena_total"] = df["cena_m2"].astype(float) * df["powierzchnia"].astype(float)
df["liczba_pokoi"] = df.liczba_pokoi.fillna(df.liczba_pokoi.mean())
df["liczba_pieter"] = df.liczba_pieter.fillna(df.liczba_pieter.mean())
df["cena_total"] = df["cena_total"].apply(lambda x: "{:.2f}".format(x))
#
#powyższe 4 są jako INT, więc użyłem wypełniania za pomocą średniej

print(df)
```
